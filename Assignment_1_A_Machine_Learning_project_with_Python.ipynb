1. Introduction
The goal of this study is to use the K-Nearest Neighbors (KNN) algorithm on both a real-world dataset (the Iris dataset) and a simulated dataset. This assignment shows how to create a basic machine learning model in Python and analyzes its performance.

2. Methodology: Iris Dataset Analysis.

 The Iris dataset was loaded utilizing sklearn.datasets.

 The dataset was divided into 80% training and 20% testing.

 A KNN classifier was trained with KNeighborsClassifier.

 The accuracy of the model was calculated with accuracy_score().

Simulated Dataset Analysis.

```python
# import modules for this project
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# load iris dataset
iris = datasets.load_iris()
data, labels = iris.data, iris.target

# training testing split
res = train_test_split(data, labels,
                       train_size=0.8,
                       test_size=0.2,
                       random_state=12)
train_data, test_data, train_labels, test_labels = res

# Create and fit a nearest-neighbor classifier
from sklearn.neighbors import KNeighborsClassifier
# classifier "out of the box", no parameters
knn = KNeighborsClassifier()
knn.fit(train_data, train_labels)

# print some interested metrics
print("Predictions from the classifier:")
learn_data_predicted = knn.predict(train_data)
print(learn_data_predicted)
print("Target values:")
print(train_labels)
print(accuracy_score(learn_data_predicted, train_labels))

# re-do KNN using some specific parameters.
knn2 = KNeighborsClassifier(algorithm='auto',
                            leaf_size=30,
                            metric='minkowski',
                            p=2,         # p=2 is equivalent to euclidian distance
                            metric_params=None,
                            n_jobs=1,
                            n_neighbors=5,
                            weights='uniform')

knn.fit(train_data, train_labels)
test_data_predicted = knn.predict(test_data)
accuracy_score(test_data_predicted, test_labels)
```

```python
# Import necessary libraries for KNN
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# Exlore the created Dataset with visualizations
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, cmap='viridis', label='Train Data', alpha=0.6)
plt.scatter(test_data[:, 0], test_data[:, 1], c=test_labels, cmap='coolwarm', marker='x', label='Test Data')
plt.title("Visualization of Simulated Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()



#  Split Data into Training and Testing Sets
train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, train_size=0.8, test_size=0.2, random_state=12
)

# Train KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(train_data, train_labels)  # Fit the model

#  Make Predictions
train_predictions = knn.predict(train_data)
test_predictions = knn.predict(test_data)

#  Calculate Accuracy
train_accuracy = accuracy_score(train_labels, train_predictions)
test_accuracy = accuracy_score(test_labels, test_predictions)

# Print results in the required format
print("Predictions from the classifier (Training Data):")
print(train_predictions)

print("Target values (Training Data):")
print(train_labels)

print(f"Training Accuracy: {train_accuracy:.6f}")

print("\nPredictions from the classifier (Testing Data):")
print(test_predictions)

print("Target values (Testing Data):")
print(test_labels)

print(f"Testing Accuracy: {test_accuracy:.6f}")

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Train Random Forest Classifier
rf = RandomForestClassifier(random_state=12)
rf.fit(train_data, train_labels)

# Train Support Vector Machine Classifier
svm = SVC(random_state=12)
svm.fit(train_data, train_labels)

# Make Predictions
rf_predictions = rf.predict(test_data)
svm_predictions = svm.predict(test_data)

# Calculate Accuracy
rf_accuracy = accuracy_score(test_labels, rf_predictions)
svm_accuracy = accuracy_score(test_labels, svm_predictions)

# Print results
print(f"Random Forest Accuracy: {rf_accuracy:.6f}")
print(f"SVM Accuracy: {svm_accuracy:.6f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# For Random Forest
print("Random Forest Confusion Matrix:")
print(confusion_matrix(test_labels, rf_predictions))
print("Random Forest Classification Report:")
print(classification_report(test_labels, rf_predictions))

# For SVM
print("SVM Confusion Matrix:")
print(confusion_matrix(test_labels, svm_predictions))
print("SVM Classification Report:")
print(classification_report(test_labels, svm_predictions))
```
